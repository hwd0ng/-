### 과제. 지니뮤직 1~200위까지 뽑기
- https://genie.co.kr/chart/top200
- https://genie.co.kr/chart/top200?ditc=D&ymd=20240520&hh=12&rtm=Y&pg=1
- https://genie.co.kr/chart/top200?ditc=D&ymd=20240520&hh=12&rtm=Y&pg=2

```python
import requests
from bs4 import BeautifulSoup
import time
```

```python
header = {'User-Agent' : 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_15_7)'}
base_url = 'https://genie.co.kr/chart/top200'
page_numbers = [1, 2, 3, 4]
results=[]
```

```python
for page in page_numbers:
    url = f'{base_url}?ditc=D&ymd=20240520&hh=12&rtm=Y&pg={page}'

    start_time = time.time()    # 시작 시간 기록

    request = requests.get(url, headers=header)
    soup = BeautifulSoup(request.text)

    titles = soup.findAll('a', {'class':'title ellipsis'})
    artists = soup.findAll('a', {'class':'artist ellipsis'})

    for i, (t, a) in enumerate(zip(titles, artists)):
        title = t.text.strip()
        artist = a.text.strip()
        results.append((i + 1 + (page - 1) * 50, title, artist))

    end_time = time.time()     # 종료 시간 기록

    total_time = end_time - start_time    # 총 소요시간

    print(f'{page}페이지 크롤링 완료! 소요시간:{total_time : .2f}초')
    print('-' * 50)  # 구분선
    time.sleep(1)

print('[ 지니뮤직 차트 1위~200위 ]')
for rank, title, artist in results:
    print(f'{rank:3d}위. {title:s} - {artist:s}')
```
